```python
### function for implementation of Perceptron



def predict(X,w):
    #computing dot product between input vector and weight vector + bias
    if X.shape[0]<len(w):
        value=np.matmul(X,w[:-1])[0]+w[-1]


    else:

        value=np.matmul(X,w)[0]


    if value>0:
        return 1
    return -1



def forward(X,y,W,bias=None,learn_rate=0.01,batch_size=1):

    # values to keep track of when an update should be made
    n_obs=0
    updates=np.zeros(W.shape) #an array to store update value

    # iteration through the whole dataset
    for i in range(len(X)):

        n_obs+=1


        if predict(X[i],W)!=y[i]:

            updates+=(learn_rate*(X[i]*y[i]).reshape(X.shape[1],1))


        if n_obs==batch_size:
            n_obs=0
            W+=updates
            updates=np.zeros(W.shape)


    W+=updates

    return W



def train_perceptron(X_train,y_train,X_val,y_val,W,epochs,bias=None,learn_rate=0.01,shuffle=False,early_stopping=0,seed=1,
                    batch_size=1, plot=True):



    if bias is not None:
        new_X_train=np.append(X_train.copy(),np.ones((X_train.shape[0],1),dtype=X_train.dtype),axis=1)
        W=np.append(W,bias)
        W=W.reshape((W.shape[0],1))
    else:
        new_X_train=np.append(X_train.copy(),np.zeros((X_train.shape[0],1),dtype=X_train.dtype),axis=1)
        W=np.append(W,0)
        W=W.reshape((W.shape[0],1))


    train_acc_list=[]
    val_acc_list=[]
    val_obs=len(y_val)
    train_obs=len(y_train)

    val_results=[]
    train_results=[]


    for i in range(len(y_train)):
        train_results.append(predict(new_X_train[i],W)==y_train[i])
    for i in range(len(y_val)):
        val_results.append(predict(X_val[i],W)==y_val[i])
    train_acc_list.append(sum(train_results)/train_obs)
    cur_val_acc=sum(val_results)/val_obs
    val_acc_list.append(cur_val_acc)

    if early_stopping>0:
        run=0
        best_val_acc=cur_val_acc
        best_weight=0
        return_weight=W
        return_train_acc=train_acc_list[-1]
        return_val_acc=cur_val_acc
    stop=False

    epoch=0

    while epoch<epochs and not stop:
        epoch+=1

        val_results=[]
        train_results=[]

        if shuffle:
            np.random.seed(seed)
            new_X_train,y_train=sklearn.utils.shuffle(new_X_train,y_train)

        W=forward(new_X_train,y_train,W,bias,learn_rate,batch_size=batch_size)


        for i in range(len(y_train)):
            train_results.append(predict(new_X_train[i],W)==y_train[i])
        for i in range(len(y_val)):
            val_results.append(predict(X_val[i],W)==y_val[i])
        train_acc_list.append(sum(train_results)/train_obs)
        cur_val_acc=sum(val_results)/val_obs
        val_acc_list.append(cur_val_acc)


        if early_stopping>0:

            if cur_val_acc>best_val_acc:
                run=0
                best_val_acc=cur_val_acc
                return_weight=W
                return_train_acc=train_acc_list[-1]
                return_val_acc=cur_val_acc

            else:

                run+=1
                stop=run>early_stopping

    if plot:
        plt.plot(range(epoch+1), train_acc_list, 'g', range(epoch+1), val_acc_list, 'r')
        plt.legend(['training accuracy','validation accuracy'])
        plt.xlabel('epochs')
        plt.ylabel('accuracy')
        plt.show()


    if early_stopping>0:
        print(f'training stop after {epoch} epochs')
        return return_weight,return_train_acc,return_val_acc

    else:
        return W,train_acc_list[-1],val_acc_list[-1]

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, precision_score, f1_score

def predict(X,w):

    if X.shape[0]<len(w):
        value=np.matmul(X,w[:-1])[0]+w[-1]


    else:

        value=np.matmul(X,w)[0]

    if value>0:
        return 1
    return -1

def forward(X, y, W, learn_rate=0.01, batch_size=1):

    updates = np.zeros(W.shape)
    n_obs = 0
    for i in range(len(X)):
        n_obs += 1
        if predict(X[i], W) != y[i]:
            updates += learn_rate * (X[i] * y[i]).reshape(W.shape)
        if n_obs == batch_size:
            W += updates
            updates = np.zeros(W.shape)
            n_obs = 0
    W += updates
    return W

def train_perceptron(X_train, y_train, X_val, y_val, W, epochs, bias=None, learn_rate=0.01, shuffle=False, early_stopping=0, seed=1, batch_size=1, plot=True):
    # Ensure that the input for training includes bias term if necessary
    if bias is not None:
        X_train = np.append(X_train, np.ones((X_train.shape[0], 1)), axis=1)
        W = np.append(W, bias).reshape(-1, 1)
    else:
        X_train = np.append(X_train, np.zeros((X_train.shape[0], 1)), axis=1)
        W = np.append(W, 0).reshape(-1, 1)

    train_acc_list = []
    val_acc_list = []

    for epoch in range(epochs):
        if shuffle:
            np.random.seed(seed)
            indices = np.random.permutation(len(X_train))
            X_train, y_train = X_train[indices], y_train[indices]

        W = forward(X_train, y_train, W, learn_rate, batch_size)

        train_predictions = [predict(X_train[i], W) for i in range(len(y_train))]
        val_predictions = [predict(np.append(X_val[i], 1), W) for i in range(len(y_val))]  # Assume bias is always last in W

        train_acc = accuracy_score(y_train, train_predictions)
        val_acc = accuracy_score(y_val, val_predictions)

        train_acc_list.append(train_acc)
        val_acc_list.append(val_acc)

        if early_stopping and len(val_acc_list) > early_stopping and val_acc <= min(val_acc_list[-early_stopping:]):
            break

    if plot:
        plt.plot(train_acc_list, label='Training Accuracy')
        plt.plot(val_acc_list, label='Validation Accuracy')
        plt.xlabel('Epoch')
        plt.ylabel('Accuracy')
        plt.legend()
        plt.show()

    # Compute validation metrics
    validation_metrics = {
        'Confusion Matrix': confusion_matrix(y_val, val_predictions),
        'Accuracy': accuracy_score(y_val, val_predictions),
        'Recall': recall_score(y_val, val_predictions, average='macro'),
        'Precision': precision_score(y_val, val_predictions, average='macro'),
        'F1 Score': f1_score(y_val, val_predictions, average='macro')
    }

    return W, train_acc_list[-1], validation_metrics

```
